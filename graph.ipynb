```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([120,70,80,90])
a=["apple","grapes","banana","cherry"]
b=["red","green","yellow","black"]
c=([0,0,.3,0])
plt.pie(x,labels=a,colors=b,explode=c,shadow=True)
plt.legend(a)
plt.show()
```
