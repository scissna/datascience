```python
#image rotaion
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('mus.jpg')
b=np.array(a)
c=ndimage.rotate(b,45,cval=110)
d=Image.fromarray(c)
d.save('mnimg.jpg')
plt.imshow(d)
```

```python
#scaling
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('mus.jpg')
b=np.array(a)
c=ndimage.zoom(b,(5,0.5,1))
d=Image.fromarray(c)
d.save('mnimg.jpg')
plt.imshow(d)
```
